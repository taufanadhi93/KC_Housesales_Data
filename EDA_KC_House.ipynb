```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kc_house_data.csv')
df.info()
df.head(10)
```

Calculating Central Tendencies and Dispersion Data of Price

```python
mn      = df['price'].min()
mx      = df['price'].max()
mean    = df['price'].mean()
mode    = df['price'].mode()
med     = df['price'].median()
std     = df['price'].std()
var     = df['price'].var()
q1      = np.quantile(df['price'],0.25)
q3      = np.quantile(df['price'],0.75)
iqr     = q3-q1
upper   = q3+(1.5*iqr)
lower   = q1-(1.5*iqr)
rg      = mx-mn
```

Calculating Central Tendencies and Dispersion Data of Grade

```python
mn2     = df['grade'].min()
mx2     = df['grade'].max()
mean2   = df['grade'].mean()
mode2   = df['grade'].mode()
med2    = df['grade'].median()
std2    = df['grade'].std()
var2    = df['grade'].var()
q12     = np.quantile(df['grade'],0.25)
q32     = np.quantile(df['grade'],0.75)
iqr2    = q32-q12
upper2  = q32+(1.5*iqr2)
lower2  = q12-(1.5*iqr2)
rg2     = mx2-mn2
print(mean2)
print(med2)
print(mode2)
```

From the histogram plot below, we can conclude that the Price distribution is Right-Skewed Distribution because the mean is greater than median. The most occurance for the price of the house is below 1 million.

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(df['price'])
ax.set_title('Price Distribution')
ax.set_xlabel('Price')
ax.set_ylabel('Frequency')

ax.axvline(x=mean, label='Mean', color='r')
ax.axvline(x=med, label='Median', color='b')
ax.legend()

plt.show()
```

From the histogram plot below, we can conclude that the Price distribution is Right-Skewed Distribution because the mean is greater than median. The most occurance for the price of the grade is grade 7.

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(df['grade'])
ax.set_title('Grade Distribution')
ax.set_xlabel('Grade')
ax.set_ylabel('Frequency')

ax.axvline(x=mean2, label='Mean', color='r')
ax.axvline(x=med2, label='Median', color='b')
ax.legend()

plt.show()
```

```python
df2 = df[['grade','sqft_lot','condition']]
df2.head(5)
```

All the variable are quantitative and relation between variable is linear, so i choose to correlation with pearson method.
Looking from the result below, grade has greatest correlation to the price. So with the increase of the grade, the price will follow to increase. If the buyer looking for the greater grade of house, the price also will be great

```python
price_pearson = pd.DataFrame(df2.corrwith(df['price'], method='pearson'),columns=['price'])
price_pearson
```

```python
priceAvg = df['price'].mean()
df3 = df[(df['price']>priceAvg)][['price','zipcode']].groupby(['zipcode']).count().sort_values('price', ascending=False).head(5)
df3
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(data=df3.reset_index(), x='zipcode', y='price')
ax.set_title(' Top 5 Zipcode with Price Greater than Average ')
ax.set_xlabel('zipcode')
ax.set_ylabel('price')
plt.show()
```

```python
df4 = df[(df['price']>priceAvg)][['price','zipcode','grade']].groupby(['zipcode']).count().sort_values('price', ascending=False)
df4
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(data=df4.reset_index(), x='zipcode', y='grade')
ax.set_title(' Bottom 5 Zipcode with the Lowest Grade')
ax.set_xlabel('zipcode')
ax.set_ylabel('grade')
plt.show()
```
